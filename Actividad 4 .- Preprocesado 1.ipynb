- Aplica en el df_b la estrategia que consideres mas adecuada para cada columna donde hagan falta datos.

```python
df_b = pd.read_csv('dirtydata.csv')
```

```python
df_b.isna().sum()
```

```python
df_b['Date']
```

```python
df_b['Date'] = df_b['Date'].replace(df_b['Date'][22], f"'2020/12/22'")
df_b['Date'] = df_b['Date'].replace(df_b['Date'][26], f"'2020/12/26'")
df_b.isna().sum()
```

```python
df_b['Calories']
```

```python
df_b['Calories'] = df_b['Calories'].replace(df_b['Calories'][18], df_b['Calories'].mean())
df_b['Calories'] = df_b['Calories'].replace(df_b['Calories'][28], df_b['Calories'].mean())
df_b.isna().sum()
```
